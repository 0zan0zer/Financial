### Nullkuponanleihe

Anleihen, welche keine Kupons zahlen

- Treasury bills, d.h. Anleihen der US-Regierung mit einer Laufzeit von maximal einem Jahr sind Nullkuponanleihen

- Vor dem Laufzeitende sind Nullkuponanleihen immer weniger wert als der Nennwert

- <b>Beispiel</b>: eine risikolose Nullkuponanleihe mit einer Laufzeit von einem Jahr und einem Nennwert von 100 EUR könnte einen Marktwert von 97,323 EUR haben.

```python
Antwort = (100/97.323)-1
Antwort
```

### Berechnung YTM bei Nullkuponanleihe

```python
def YTM_Nullkupon(N,P,n):
    return print("(("+str(N)+"/"+str(P)+")^(1/"+str(n)+"))-1= "+str((((N/P)**(1/n))-1)))
```

```python
YTM_Nullkupon(100,97.65,8)
```

- Ist die periodisierte Rendite, die man erzielt, wenn man die Anliehe bis zum Ende der Laufzeit hält

- YTM von risikolosen Nullkuponanleihen (z.B. in der Praxis oft Staatsanleihen) werden oft auch als „risikolose Rendite“ bezeichnet (oder auch als Spotzins)

### Kuponanleihen

Kuponanleihen zahlen dem Investor
- Den Nennwert zum Laufzeitende
- Regelmäßige Kuponzahlungen (Zinszahlungen) zwischen Emission und Laufzeitende

<b>Beispiel:</b> am 18. Mai 2018 hat das Bundesfinanzministerium eine Anleihe mit einer Laufzeit bis zum Mai 2023, einem Nennwert von 20.000 EUR und einer Kuponrate von 1.5% emittiert. Kupons werden halbjährlich gezahlt. Die erste
Kuponzahlung ist zum 18. November 2018 fällig.

```python
def show_kupon(K,P,n):
    return print("("+str(K)+"*("+str(P)+"/"+str(n)+")= "+str((K*(P/n))))
def kupon(K,P,n):
    return (K*(P/n))
```

### Preis einer Kuponanleihe

```python
def show_kuponanleihe(K,N,i,n):
    return print("("+str(K)+"/"+str(i)+")*(1-(1/((1+"+str(i)+")^"+str(n)+")))+("+str(N)+"/(1+"+str(i)+")^"+str(n)+")= "+str((K/i)*(1-(1/((1+i)**n)))+(N/(1+i)**n)))
def kuponanleihe(K,N,i,n):
    return (K/i)*(1-(1/((1+i)**n)))+(N/(1+i)**n)
```

```python
kuponanleihe(2,100,0.03,5)
```
